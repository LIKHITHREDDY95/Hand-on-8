```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


class SinglyLinkedList:
    def __init__(self, capacity):
        self.capacity = capacity
        self.nodes = [None] * capacity
        self.head = None
        self.size = 0
        self.free_list = list(range(capacity))

    def insert_at_beginning(self, data):
        if self.is_full():
            raise Exception("List is full")
        new_node_index = self.free_list.pop()
        self.nodes[new_node_index] = Node(data)
        self.nodes[new_node_index].next = self.head
        self.head = new_node_index
        self.size += 1

    def insert_at_end(self, data):
        if self.is_full():
            raise Exception("List is full")
        new_node_index = self.free_list.pop()
        self.nodes[new_node_index] = Node(data)
        if self.is_empty():
            self.head = new_node_index
        else:
            current = self.head
            while self.nodes[current].next is not None:
                current = self.nodes[current].next
            self.nodes[current].next = new_node_index
        self.size += 1

    def delete_from_beginning(self):
        if self.is_empty():
            raise Exception("List is empty")
        deleted_node_index = self.head
        self.head = self.nodes[self.head].next
        self.nodes[deleted_node_index] = None
        self.free_list.append(deleted_node_index)
        self.size -= 1

    def delete_from_end(self):
        if self.is_empty():
            raise Exception("List is empty")
        if self.size == 1:
            self.delete_from_beginning()
        else:
            current = self.head
            while self.nodes[self.nodes[current].next].next is not None:
                current = self.nodes[current].next
            deleted_node_index = self.nodes[current].next
            self.nodes[current].next = None
            self.nodes[deleted_node_index] = None
            self.free_list.append(deleted_node_index)
            self.size -= 1

    def is_empty(self):
        return self.size == 0

    def is_full(self):
        return self.size == self.capacity

    def print_list(self):
        current = self.head
        while current is not None:
            print(self.nodes[current].data, end=" -> ")
            current = self.nodes[current].next
        print("None")

if __name__ == "__main__":
    linked_list = SinglyLinkedList(5)

    print("Inserting elements at the beginning:")
    for i in range(1, 4):
        linked_list.insert_at_beginning(i)
        print(f"Inserted {i} at the beginning")
        linked_list.print_list()

    print("\nInserting elements at the end:")
    linked_list.insert_at_end(4)
    print("Inserted 4 at the end")
    linked_list.print_list()
    linked_list.insert_at_end(5)
    print("Inserted 5 at the end")
    linked_list.print_list()

    print("\nDeleting from the beginning:")
    linked_list.delete_from_beginning()
    linked_list.print_list()

    print("\nDeleting from the end:")
    linked_list.delete_from_end()
    linked_list.print_list()

    try:
        linked_list.insert_at_end(6)
    except Exception as e:
        print(f"\nError: {e}")

    print("\nDeleting all elements:")
    while not linked_list.is_empty():
        linked_list.delete_from_beginning()
        linked_list.print_list()

    try:
        linked_list.delete_from_end()
    except Exception as e:
        print(f"\nError: {e}")
```
